```python
class tableDataset(Dataset):
    def __init__(self, data, target, domain_id= 0 ,mode = 'train', train_split_ratio = 0.7):
        super(tableDataset, self).__init__()
        split_ratio = int(data.shape[0])
        if mode == 'train':
            self.data = data[:split_ratio]
            self.target = target[:split_ratio]
        else:
            self.data = data[split_ratio:]
            self.target = target[split_ratio:]
        self.domain_id =domain_id
        
            

    def __getitem__(self, index):
        x = self.data[index]
        y = self.target[index]
        return x, y, self.domain_id

    def __len__(self):
        return self.data.shape[0]
```
